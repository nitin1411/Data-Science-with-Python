```python
import pandas as pd
import matplotlib.pyplot as plt
tc_data = pd.read_csv("C:\\Users\\nitin\\Desktop\\Assignments\\Multiple Linear Regression\\ToyotaCorolla.csv", encoding= 'unicode_escape')
```

```python
tc_data.head(10)
```

```python
tc_data.columns
```

```python
tc = tc_data[['Price','Age_08_04','KM','HP','cc','Doors','Gears','Quarterly_Tax','Weight']]
```

```python
tc
```

```python
tc.corr() #correlation matrix
```

```python
import seaborn as sns
plt.subplots(figsize=(10, 6))
correlationMatrix = tc.corr().abs()
sns.heatmap(correlationMatrix,annot=True)
plt.show()
```

```python
# preparing model considering all the variables 
import statsmodels.formula.api as smf 
         
# Preparing model                  
tc_mod1 = smf.ols('Price~Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=tc).fit() 
```

```python
# Getting coefficients of variables               
tc_mod1.params
```

```python
# Model Summary
tc_mod1.summary()
```

```python
# p-values for cc,Doors are more than 0.05 but we know that [cc,Doors] has no high correlation value
#preparing model without 'cc' and 'doors'
tc_mod2 = smf.ols('Price~Age_08_04+KM+HP+Gears+Quarterly_Tax+Weight',data=tc).fit() 
tc_mod2.summary()
```

```python
    # there is no any significant changes in r-sq and adj r-sq value by developing model without "cc and Doors"
    #we'll try and remove the outliers(if any)
```

```python
import statsmodels.api as sm
sm.graphics.influence_plot(tc_mod1)
```

```python
tc2 =tc.drop(tc.index[[80,960,221,660]],axis=0)
```

```python
# Preparing model after iteration              
tc_mod3 = smf.ols('Price~Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=tc2).fit() 
tc_mod3.summary()
```

```python
#checking again for outliers
sm.graphics.influence_plot(tc_mod3)
```

```python
tc3 = tc2.drop(tc2.index[[601,654,991]], axis = 0)
```

```python
# Preparing model after second iteration                
tc_mod4 = smf.ols('Price~Age_08_04+KM+HP+cc+Doors+Gears+Quarterly_Tax+Weight',data=tc3).fit() 
tc_mod4.summary()
```

```python
#with R-squared and Adj.R-squared same at 0.885, we have built a good model.
```
